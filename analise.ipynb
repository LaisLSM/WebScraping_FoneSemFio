```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar seu dataframe tratado
df = pd.read_excel('Fones sem Fio.xlsx')

# Configurações básicas
plt.style.use('seaborn-v0_8-darkgrid')  # Visual bonito e tradicional
sns.set_palette('pastel')  # Cores suaves
```

```python
# Fone mais caro
fone_mais_caro = df.loc[df['Preço'].idxmax()]
print("Fone mais caro:")
print(f"Produto: {fone_mais_caro['Descrição do Produto']}")
print(f"Preço: R$ {fone_mais_caro['Preço']:.2f}")
```

```python
# Quantidade de fones por cor

plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Cor', order=df['Cor'].value_counts().index)
plt.title('Quantidade de Fones por Cor')
plt.xlabel('Quantidade')
plt.ylabel('Cor')
plt.show()
```

```python
# Gráfico de quantidade de fones por marca
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Marca', order=df['Marca'].value_counts().index)
plt.title('Quantidade de Fones por Marca')
plt.xlabel('Quantidade')
plt.ylabel('Marca')
plt.show()
```

```python
# Quais os fones mais caros (top 5)

top5_caros = df.sort_values(by='Preço', ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(data=top5_caros, x='Preço', y='Descrição do Produto', palette='muted')
plt.title('Top 5 Fones Mais Caros')
plt.xlabel('Preço (R$)')
plt.ylabel('Produto')
plt.show()
```
